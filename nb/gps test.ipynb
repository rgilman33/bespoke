```python
import sys
sys.path.append('/home/beans/bespoke')
from serial import Serial
import io
from constants import *
import pynmea2

TIMEOUT = 1. / GPS_HZ
SERIAL_PORT = '/dev/ttyUSB0'
baud_rate = 115200 #4800
ser = Serial(SERIAL_PORT, baud_rate, timeout=TIMEOUT, )
ser_io = io.TextIOWrapper(io.BufferedRWPair(ser, ser)) # from the pynmea2 docs
```

```python
while True:
    try:
        line = ser_io.readline()
    except UnicodeDecodeError as e:
        print(f"readline failure:{e}")
        continue
    try:
        msg = pynmea2.parse(line)
        if msg.identifier()=='GPGGA,':        
            print(msg.num_sats, msg.gps_qual, msg.age_gps_data, msg.latitude, msg.longitude, msg.timestamp)
    except:
        pass
```
