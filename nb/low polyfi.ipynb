```python
import shutil, subprocess, glob, time
```

```python
car_blends = glob.glob("/home/beans/Downloads/Tranportation_Pro_v4/Tranportation/data/vehicles/*/*/*lowpoly.blend")
no_include = ["Bicycle", "Plane", "Boat", "Tractor", "Train", "Tramway"]
car_blends = [f for f in car_blends if len([n for n in no_include if n in f])==0]

for f in car_blends:
    adj_path = f.replace("lowpoly.blend", "lowpoly_ADJUSTED.blend")
    shutil.copyfile(f, adj_path)
    print(adj_path)
    subprocess.Popen(f"(cd /home/beans/blender_3.3; ./blender '{adj_path}' --background --python /home/beans/bespoke/datagen/low_polyfi.py)", shell=True)
    time.sleep(.5)
```
